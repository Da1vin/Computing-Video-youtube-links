```python
def merge_sort(numbers, start, end): # start & end are indicators of the region we are sorting

    # base case: start >= end, this means 1 or no more items left within the region between start and end

    if start < end:

        ### recursively split and sort left & right halves
        # the changing parameters are start or end, numbers list remain exactly the same length throughout

        mid = (start + end) // 2 # calculate midpoint of the region to split
        merge_sort(numbers, start, mid) # left half (up till mid) will be sorted after calling this procedure
        merge_sort(numbers, mid+1, end) # right half (mid+1 till end) will be sorted after calling this procedure

        ### merge left and right together (we'll just put the algo in the same procedure)

        results = [] # to accumulate the sequence of sorted items

        left_index = start # this keeps track of which item in left half we're currently comparing
        right_index = mid+1 # this keeps track of which item in right half we're currently comparing

        # loop for as long as we still have items left to compare in left & right half
        while left_index <= mid and right_index <= end:
            if numbers[left_index] < numbers[right_index]: # assume ascending sort
                results.append(numbers[left_index])
                left_index += 1
            else:
                results.append(numbers[right_index])
                right_index += 1

        # append leftover elements from left (if any)
        while left_index <= mid:
            results.append(numbers[left_index])
            left_index += 1

        # append leftover elements from right (if any)
        while right_index <= end:
            results.append(numbers[right_index])
            right_index += 1

        # replace the items in numbers list within the start & end region with the sorted items
        index = 0 # keep track of which item we're inserting back into numbers list
        for i in range(start, end+1): # inclusive of end
            numbers[i] = results[index]
            index += 1
        return numbers   
# main
data = [4,5,1,9,7,3]
print(merge_sort(data, 0, len(data)-1))
```
